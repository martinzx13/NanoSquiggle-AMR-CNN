
  Lecture: What is a 1D-CNN?


  Let's break this down from first principles. You understand that Linear Regression computes a weighted sum of inputs. A Convolutional Neural Network (CNN) does the same, but in a
  localized, sliding-window fashion.


  Imagine a small vector of weights, called a kernel or filter (e.g., [w1, w2, w3]). Instead of applying it to all your input data at once, you slide this kernel across your 1D signal (the
  Nanopore squiggle):
   1. You compute the dot product of the kernel with the first three points of your signal. That's your first output value.
   2. You slide the kernel one step to the right and compute the dot product with points 2, 3, and 4 of your signal. That's your second output value.
   3. You repeat this across the entire signal.


  The output is a new, transformed signal representing how strongly the feature defined by your kernel (e.g., a specific shape or pattern) was detected at each position. A 1D-CNN learns
  dozens of these kernels automatically via backpropagation to detect the specific squiggle patterns that are predictive of antibiotic resistance. It is a hierarchical feature detector for
  sequential data.
