  Lecture: The Path to the Real World (Stage 3)

  You have mastered the Architect phase for synthetic signals. But your research project
  isn't about Sine waves; it's about Nanopore squiggles.

  In the real world:
   1. Variable Lengths: Nanopore signals aren't always exactly 500 points. You will need to
      handle padding or use Global Average Pooling to ensure your classifier works on any
      length.
   2. Raw Signal Normalization: Real electrical signals have "drift" and "offset." You will
      need to apply Z-score normalization (subtract mean, divide by std) to every squiggle
      before it hits the model.
   3. Data Imbalance: In AMR prediction, you might have 1,000 "Susceptible" signals and
      only 50 "Resistant" signals. Your model will try to cheat by always predicting
      "Susceptible."
